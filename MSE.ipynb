```python
import numpy as np
import matplotlib.pyplot as plt
```

Mean square error; 
$$ MSE = \frac{1}{N} \sum_1^N(x_i - \bar{x})^2 $$
Where N is number of samples from some distribution. <br />
$x_i$ is mean of each sample <br />
$\bar{x}$ is real value of estimated parameter <br /> 
Here: Taking samples independenty from Gaussian distribution and testing two variance approximators

```python
def varEst(x):
    Npoints = len(x)
    sum = 0
    for i in x:
        sum += (i - x.mean())**2
    return (sum/Npoints, sum/(Npoints+1))

def MeanSquareError(realParam, Ntests, mu = 0, var = 1, Npoints = 1000):
    MSE_bias = 0
    MSE_unbiased = 0
    for i in range(Ntests):
        x = np.random.normal(mu, var, Npoints)
        BiasedEstimator, UnbiasedEstimator = varEst(x)
        MSE_bias +=(BiasedEstimator - realParam)**2
        MSE_unbiased =+(UnbiasedEstimator - realParam)**2
    return MSE_bias/Ntests, MSE_unbiased/Ntests
def Plot(N):
    biased = np.zeros(N)
    unbiased = np.zeros(N)
    
    tt = np.linspace(0, 100, N)
    for i in range(1, N):
        biasedEst, unbiasedEst = MeanSquareError(1, i)
        biased[i] = biasedEst
        unbiased[i] = unbiasedEst
    fig, ax = plt.subplots(1,1, figsize =(12, 5))
    plt.plot(tt, biased)
    plt.title(r'Mean Square Error function of Biased Estimator of $\sigma^2$  $\frac{1}{n} \sum_1^n(x_i - \bar{x})^2$')
    fig, ax = plt.subplots(1,1, figsize =(12, 5))
    plt.plot(tt, unbiased)
    plt.title(r'Mean Square Error function of Unbiased Estimator of $\sigma^2$  $\frac{1}{n+1} \sum_1^n(x_i - \bar{x})^2$')
    plt.show()
Plot(100)
```
